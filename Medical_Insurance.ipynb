# **Data Preprocessing**

In the information section we saw that there are 3 object type features,and for the model we have to encode it 

```python
genders = {"male": 0, "female": 1}
df['sex'] = df['sex'].map(genders)
df.head()
```

encoding sex column, where female will be denoted as 1 and male will be denoted as 0

```python
Smoker={"yes":1,"no":0}
df['smoker']= df['smoker'].map(Smoker)
df
```

Whether the person is smoker or not 1 will represent "yes" and 0 will denote as "no"

```python
Region = {"southwest": 0, "southeast": 1,"northeast":2,"northwest":3}
df['region'] = df['region'].map(Region)
df.head()
```

changing or encoding the names of the region like 'southwest','southeast','northeast','northwest' into 0,1,2,3 respectively

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, linewidths=0.5, fmt= '.3f')
```

```python
df.info()
```

Now everything looks fine. So let's start model building.

# **Building the model**

Firstly we have to split the dataset into train set and test set.

```python
from sklearn.model_selection import train_test_split
x=df.drop('charges',axis=1)
y=df['charges']
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.30)
```

Checking Dimensions of Train and Test

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

Before moving ahead, let's import all the models from sklearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

# Linear Regression

```python
lr = LinearRegression() 
lr=lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
pred = lr.predict(X_test)
s= mean_absolute_error(y_test,pred)
s1= mean_squared_error(y_test,pred)
s2= r2_score(y_test,pred)

print("The MAE with the linear regressor is: "+str(s))
print("The MsE with the linear regressor is: "+str(s1))
print("The R2_Score with the linear regressor is: "+str(s2))
```

# Decision Tree Regressor

```python
dtr = DecisionTreeRegressor()
dtr=dtr.fit(X_train,y_train)
```

```python
pred = dtr.predict(X_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)

print("The MAE with the DT regressor is: "+str(s))
print("The MsE with the DT regressor is: "+str(s1))
print("The R2_Score with the DT regressor is: "+str(s2))
```

##Random Forest Regressor

```python
r = RandomForestRegressor() #create the object of the model
r=r.fit(X_train,y_train)
```

```python
pred = r.predict(X_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)

print("The MAE with the RF regressor is: "+str(s))
print("The MsE with the RF regressor is: "+str(s1))
print("The R2_Score with the RF regressor is: "+str(s2))
```

# K-Neighbors Regressor

```python
knn=KNeighborsRegressor()
knn=knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)

print("The MAE with the KNN regressor is: "+str(s))
print("The MsE with the KNN regressor is: "+str(s1))
print("The R2_Score with the KNN regressor is: "+str(s2))
```

# XGB Regressor

```python
xgb=XGBRegressor()
xgb=xgb.fit(X_train,y_train)
```

```python
pred = xgb.predict(X_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)

print("The MAE with the XGB regressor is: "+str(s))
print("The MsE with the XGB regressor is: "+str(s1))
print("The R2_Score with the XGB regressor is: "+str(s2))
```
